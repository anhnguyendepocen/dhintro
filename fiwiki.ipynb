```r
# Load fiwiki inflection frequency file
fiwiki <- read.csv(gzfile("/srv/data/fiwiki_cases.tsv.gz"),sep="\t")
head(fiwiki)
```

```r
# Take the inflection forms (but not the frequency) for the 100 most frequent terms, associate them to the term.
fiwiki2 <- data.frame(fiwiki[0:100,c(3:29)],row.names=fiwiki[0:100,2])
head(fiwiki2)
```

```r
# Cluster the terms
options(repr.plot.width=10, repr.plot.height=8)
plot(hclust(dist(fiwiki2)))
```
